```python
import pandas as pd
import matplotlib.pyplot as plt

# read the csv files and plot the data
df_AT = pd.read_csv(r"D:\VScode\van2015_repeat\data\AT.csv")
df_CO2 = pd.read_csv(r"D:\VScode\van2015_repeat\data\CO2.csv")
df_CH4 = pd.read_csv(r"D:\VScode\van2015_repeat\data\CH4.csv")
df_inso = pd.read_csv(r"D:\VScode\van2015_repeat\data\insolation.csv")

# crop the data and get the age>=0 and <=360000
age_start=360000
age_end=799000

df_AT = df_AT[(df_AT['age'] >= age_start) & (df_AT['age'] <= age_end)]
df_CO2 = df_CO2[(df_CO2['age'] >= age_start) & (df_CO2['age'] <= age_end)]
df_CH4 = df_CH4[(df_CH4['age'] >= age_start) & (df_CH4['age'] <= age_end)]
df_inso = df_inso[(df_inso['age'] >= age_start) & (df_inso['age'] <= age_end)]

fig, axs = plt.subplots(4, 1, figsize=(10, 15))

axs[0].plot(df_AT['age'], df_AT['AT'], color='r')
axs[0].set_title('AT')
axs[0].set_xlabel('age')
axs[0].set_ylabel('AT')

axs[1].plot(df_CO2['age'], df_CO2['CO2'], color='g')
axs[1].set_title('CO2')
axs[1].set_xlabel('age')
axs[1].set_ylabel('CO2')

axs[2].plot(df_CH4['age'], df_CH4['CH4'], color='b')
axs[2].set_title('CH4')
axs[2].set_xlabel('age')
axs[2].set_ylabel('CH4')

axs[3].plot(df_inso['age'], df_inso['inso'], color='k')
axs[3].set_title('insolation')
axs[3].set_xlabel('age')
axs[3].set_ylabel('insolation')

plt.tight_layout()
plt.show()

# print the length of the data
print(len(df_AT['age']))
```
